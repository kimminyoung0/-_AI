### 모델 학습 (위도 예측)

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import catboost
from scipy.stats import randint, loguniform
import time

# 데이터를 학습용과 테스트용으로 나누기
Train_x, Test_x, Train_y, Test_y = train_test_split(X_train, Y_train_la, 
                                                    test_size=0.3, 
                                                    random_state=37)

param_distribs = {
    'n_estimators': randint(100, 300),
    'depth': randint(1, 5),
    'learning_rate': loguniform(1e-3, 0.1),
    'min_child_samples': randint(10, 40),
    'grow_policy': ['SymmetricTree', 'Lossguide', 'Depthwise']
}

start = time.time()

# CatBoost 모델 초기화
modelcat_la = catboost.CatBoostRegressor(eval_metric='R2')

# 랜덤 서치를 통한 하이퍼파라미터 튜닝
modelcat_la.randomized_search(param_distribs,
                              Train_x, Train_y,
                              cv=3, n_iter=30)

# 있어도 되고 없어도 됨. 위에서 학습이 됨.
#modelcat_la.fit(Train_x, Train_y)

# 학습된 모델을 사용하여 예측 학습데이터에서 분리되어 나온 test 데이터로 하는 것.
preds= modelcat_la.predict(Test_x)

# R2 계수
r2 = r2_score(Test_y, preds)

# 평균제곱오차(MSE)
mse = mean_squared_error(Test_y, preds)

end = time.time()
print(f'Time elapsed: {end - start}')


print(f'R2 score on training set: {r2}')
print(f'Mean Squared Error on test set: {mse}')
```

### 모델학습 (경도 예측)

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import catboost
from scipy.stats import randint, loguniform
import time

# 데이터를 학습용과 테스트용으로 나누기
Train_x, Test_x, Train_y, Test_y = train_test_split(X_train, Y_train_lo, 
                                                    test_size=0.3, 
                                                    random_state=37)

param_distribs = {
    'n_estimators': randint(100, 300),
    'depth': randint(1, 5),
    'learning_rate': loguniform(1e-3, 0.1),
    'min_child_samples': randint(10, 40),
    'grow_policy': ['SymmetricTree', 'Lossguide', 'Depthwise']
}

start = time.time()

# CatBoost 모델 초기화
modelcat_lo = catboost.CatBoostRegressor(eval_metric='R2')

# 랜덤 서치를 통한 하이퍼파라미터 튜닝
modelcat_lo.randomized_search(param_distribs,
                              Train_x, Train_y,
                              cv=3, n_iter=30)

# 있어도 되고 없어도 됨. 위에서 학습이 됨.
#modelcat_la.fit(Train_x, Train_y)

# 학습된 모델을 사용하여 예측 학습데이터에서 분리되어 나온 test 데이터로 하는 것.
preds= modelcat_lo.predict(Test_x)

# R2 계수
r22 = r2_score(Test_y, preds)

# 평균제곱오차(MSE)
mse2 = mean_squared_error(Test_y, preds)

end = time.time()
print(f'Time elapsed: {end - start}')


print(f'R2 score on training set: {r22}')
print(f'Mean Squared Error on test set: {mse2}')
```
